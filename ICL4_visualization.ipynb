```python
def get_data(state='active'):
    if state=='active':
        file = 'valid_a.csv'
    if state=='inactive':
        file = 'valid_i.csv'
    data = pd.read_csv(file, dtype = {'PDB': str})
    data.target_angles = data.target_angles.apply(literal_eval)
    return data
```

```python
def get_interactions(active, inactive):
    # for each active lookup the proteins with same genes and calculate delta_phi_psi
    cols = ['query_name', 'query_gene', 'query_seq', 'query_angles', 'inactive_states', 'n_inactive_states']
    interactions = pd.DataFrame(columns=cols)
    for i in range(len(active)):
        query_name = active.iloc[i]['PDB']
        query_gene = active.iloc[i]['uniprot(gene)']
        query_angles = active.iloc[i]['target_angles']
        query_seq = active.iloc[i]['TM7_aligned_seg']
        inactive_states = inactive[inactive['uniprot(gene)'] == query_gene][['PDB', 'target_angles', 'TM7_aligned_seg']].reset_index().to_dict()
        n_inactive_states = len(inactive_states['index'])
        interactions.loc[i] = [query_name, query_gene, query_seq, query_angles, inactive_states, n_inactive_states]
    return interactions.reset_index(drop=True)
```

```python
seq_l = []
names_i_i = []
names_a_i = []
genes_i = []
angle_diffs_i = []
for i in range(len(valid_interactions)):
    name_a = valid_interactions.iloc[i]['query_name']
    gene_a = valid_interactions.iloc[i]['query_gene']
    genes_i.append(gene_a)
    angles_a = np.asarray(valid_interactions.iloc[i]['query_angles'])
    angles_i = pd.DataFrame(valid_interactions['inactive_states'].iloc[i]['target_angles']).T
    query_seq = valid_interactions['query_seq'].iloc[i]
    names_j = []
    angle_diffs_j = []
    for j in range(len(angles_i)):
        name_j = valid_interactions['inactive_states'].iloc[i]['PDB'][j]
        names_j.append(name_j)
        angles_j = pd.DataFrame(angles_i).iloc[j]
        angles = []
        for k in range(len(angles_j)):
            angles.append(np.asarray(angles_j.iloc[k]))
        angles_j_ = np.asarray(np.vstack(angles))
        max_len = min(angles_a.shape[0], angles_j_.shape[0])  # this should be fixed, namely -> we only use the correct angles
        angle_diff = angles_a[:max_len,:] - angles_j_[:max_len, :]
        angle_diffs_j.append(angle_diff)
    seq_l.append(query_seq[:max_len])
    names_i_i.append(names_j)
    names_a_i.append(name_a)
    angle_diffs_i.append(np.asarray(angle_diffs_j))
```

```python
def plot_tm7(pdb, gene, data, angle='psi'):
    #plt.figure(figsize=(20,10))
    data.plot(figsize=(20,10))
    plt.title('Delta '+angle+' - ' +gene+': '+pdb)
    plt.legend(loc='best')
```
