```python
from Processing.processor import *
```

This should be run on a cluster with a run file creating multiple runs in parallel and then we combine everything into a single dataframe that is then loaded into our chem-env!


```python
def create_multiple(n=1, limit=2):
    for i in range(n):
        p = PDBProcessor(parse=True, start = i*limit, limit=limit, setting=3, verbose=0)
        p.create_training_df()
```

```python
n = 10
limit = 3

create_multiple(n, limit)
```

```python
p = PDBProcessor(parse=False, setting=3, verbose=0)
p.create_complete_trainingset(step=limit, limit=n*limit)
```
