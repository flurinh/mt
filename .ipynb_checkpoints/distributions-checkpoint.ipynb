Ref: https://proteinstructures.com/Structure/Structure/Ramachandran-plot.html Consider point cloud (points) as as vectors. We now want to fit a Kernel Density Estimation Model from which we can sample from: http://www.cs.columbia.edu/~jebara/papers/Kondor,Jebara_point_set.pdf or https://link.springer.com/chapter/10.1007/11564096_27

```python
from Processing.distribution import *
```

```python
from Processing.loader import *

data = DataSet(limit=20)
```

### Lets inspect TRANS-Amidplane angle plot

```python
ad = AngleDistr(psi=data.data.psi, 
                phi=data.data.phi, 
                omega=data.data.omega, 
                batch_size=120,
                norm=False,
                log=True)
ad.kde()
ad.plot()
```

```python
samples = ad.sample()
```

```python
ad.plot_samples(samples)
```

### Let's inspect (the very sparse) CIS-Amidplane angle plot

```python
ad = AngleDistr(psi=data.data.psi, 
                phi=data.data.phi, 
                omega=data.data.omega, 
                batch_size=120,
                norm=False,
                log=True,
                mode='cis')
ad.kde()
ad.plot()
```
